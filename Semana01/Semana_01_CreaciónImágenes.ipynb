```python
width, height = 500, 500
image = Image.new("RGB", (width, height))
draw = ImageDraw.Draw(image)
colors = ["#000000", "#fe380d", "#98fe0d", "#150dfe", "#000000"]
border = 90
thickness = 30

for color in colors:
    draw.rectangle([border, border, width - border, height - border], outline=color, fill=color)
    border += thickness

image.show()
```

Lo m√°s cercano que se logr√≥ fue la siguiente imagen:

```python
# Crear una figura y un eje
fig, ax = plt.subplots()

# Generar una malla de coordenadas m√°s densa
x = np.linspace(0, 50 * np.pi, 350)  # Aumentar la cantidad de ciclos y puntos para m√°s ondas
y = np.linspace(0, 50 * np.pi, 1350)
X, Y = np.meshgrid(x, y)

# Generar ondas sinusoidales separadas en X y Y
Z = np.sin(16 * X) + np.sin(1*Y)  # Ondas independientes en cada eje

# Aumentar el contraste de las ondas
Z = np.sign(Z)  # Convertir los valores a -1 y 1 para que las ondas sean m√°s marcadas

# Invertir los valores para que las ondas sean blancas sobre fondo negro
Z = (Z + 1) / 2  # Normalizar entre 0 y 1 para aplicar el color

# Crear un mapa de colores personalizado (negro y blanco)
cmap = mcolors.ListedColormap(['black', 'white'])

# Mostrar la imagen con un fondo negro y las ondas blancas
ax.imshow(Z, cmap=cmap, interpolation='none', extent=[0, 50*np.pi, 0, 50*np.pi])

# Quitar los ejes
ax.axis('off')

# Cambiar el color de fondo a negro
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Guardar la imagen
plt.savefig("imagen_ondas_blanco_negro.png", bbox_inches='tight', pad_inches=0, dpi=300, facecolor='black')
plt.show()
```
